# Particle Swarm Optimization

In chapter 1 we're introduced to the core mechanics of PSO, and how we can create particles.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Exercise 1

```python
def create_particles(num_particles, search_space):
    return np.random.uniform(search_space[0], search_space[1], size=(num_particles, 2))

particles = create_particles(200, [-10, 10])
```

Exercise 2

```python
particles_x = particles[:, 0]
particles_y = particles[:, 1]

fig, ax = plt.subplots()
ax.scatter(particles_x, particles_y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.title("Basic particles")
plt.show()
```

Exercise 3


As we increase the particles, we can see that it starts filling up the search space. Of course, as the search space increases, the spread of the particles also increases.

The initial distribution of the particles is important, as it could mean the difference between (quickly) finding the best optima, or not finding it at all.
